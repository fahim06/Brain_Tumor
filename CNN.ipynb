```python
import numpy as np
import pandas as pd
import sklearn.metrics
import matplotlib.pyplot as plt
```

```python
trainingDataset=pd.read_csv("/content/drive/MyDrive/Defence/Code/Brain_Tumour/Training.csv")
```

```python
testingDataset=pd.read_csv("/content/drive/MyDrive/Defence/Code/Brain_Tumour/Testing.csv")
```

```python
training_image=[]
```

```python
trainingDataset['Image'][2]
```

```python
from tqdm import tqdm
from keras.preprocessing import image
for i in tqdm(range(trainingDataset.shape[0])):
  print(trainingDataset['Image'][i])
  img=image.load_img("/content/drive/MyDrive/Defence/Code/Brain_Tumour/Training/"+trainingDataset['Image'][i],target_size=(256,256,3),grayscale=False)
  img=image.img_to_array(img)
  training_image.append(img)
```

```python
X_train=np.array(training_image)
```

```python
X_train.shape
```

```python
X_train=X_train.astype("float32")/255
```

```python
y_train=trainingDataset['Class'].values
```

```python
y_train
```

```python
y_train.shape
```

```python
y_train=np.reshape(y_train, (-1, 1))
```

```python
y_train.shape
```

```python
from keras.utils import np_utils
from keras import backend as K
from keras.models import Sequential
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.layers.core import Activation, Flatten, Dense, Dropout
```

```python
Y_train = np_utils.to_categorical(y_train, 2)
```

```python
Y_train
```

# **CNN**

```python
IMG_ROWS=256
IMG_COLS=256
IMG_CHANNELS=3
```

```python
NB_EPOCH=20
BATCH_SIZE=15
VERBOSE=1
OPTIMIZER='SGD'
VALIDATION_SPLIT=0.2
NB_CLASSES=2
```

```python
model=Sequential()
model.add(Conv2D(32,kernel_size=2,padding='same',input_shape=(IMG_ROWS,IMG_COLS,IMG_CHANNELS)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
model.add(Dropout(0.25))
model.add(Conv2D(50,kernel_size=2,padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.2))
model.add(Dense(NB_CLASSES))
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy", optimizer=OPTIMIZER, metrics=["accuracy"])
```

```python
history = model.fit(X_train, Y_train, batch_size=BATCH_SIZE, epochs=30,verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
```

```python
# list all data in history
print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
